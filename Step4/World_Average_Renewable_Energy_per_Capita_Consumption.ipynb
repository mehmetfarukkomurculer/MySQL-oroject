```python
cnx = mysql.connector.connect(
    host='localhost',
    user='root',
    password='password',
    database='cs306proj'
)
```

```python
cursor = cnx.cursor()
```

```python
query = "SELECT * FROM per_capita_consumption"
cursor.execute(query)
result = cursor.fetchall()
```

```python
df = pd.DataFrame(result, columns=[column[0] for column in cursor.description])
df = df[df['year'] <= 2020]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.lineplot(x='year', y='renewables_per_capita', data=df, ax=ax)
ax.set_title('World Average Renewable Energy per Capita Consumption')
```
